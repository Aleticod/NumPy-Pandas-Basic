# Concat, merge y join

```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A':['A0', 'A1', 'A2', 'A3'],
'B':['B0', 'B1', 'B2', 'B3'],
'C':['C0', 'C1', 'C2', 'C3'],
'D':['D0', 'D1', 'D2', 'D3']})
```

```python
df2 = pd.DataFrame({'A':['A4', 'A5', 'A6', 'A7'],
'B':['B4', 'B5', 'B6', 'B7'],
'C':['C4', 'C5', 'C6', 'C7'],
'D':['D4', 'D5', 'D6', 'D7']})
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
pd.concat([df1,df2], axis=1)
```

## Merge

```python
izq = pd.DataFrame({'key': ['k0','k1', 'k2', 'k3'],
'A':['A0', 'A1', 'A2', 'A3'],
'B':['B0', 'B1', 'B2', 'B3']})
```

```python
izq
```

```python
der = pd.DataFrame({'key': ['k0','k1', 'k2', 'k3'],
'C':['C0', 'C1', 'C2', 'C3'],
'D':['D0', 'D1', 'D2', 'D3']})
```

```python
der
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key': ['k0','k2', 'k3', 'k5'],
'A':['A0', 'A1', 'A2', 'A3'],
'B':['B0', 'B1', 'B2', 'B3']})

der = pd.DataFrame({'key_2': ['k0','k1', 'k3', 'k4'],
'C':['C0', 'C1', 'C2', 'C3'],
'D':['D0', 'D1', 'D2', 'D3']})
```

```python
izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='left')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='inner')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='outer')
```

## Join: index match

```python
iz = pd.DataFrame({'A':['A0', 'A1', 'A2'],
'B':['B0', 'B1', 'B2']}, index=['k0', 'k1', 'k2'])
```

```python
iz
```

```python
de = pd.DataFrame({'C':['C0', 'C1', 'C2'],
'D':['D0', 'D1', 'D2']}, index = ['k0','k2', 'k3'],)
```

```python
de
```

```python
iz.join(de)
```

```python
iz.join(de, how = 'inner')
```

```python
iz.join(de, how = 'left')
```

```python
iz.join(de, how = 'right')
```

```python
iz.join(de, how = 'outer')
```
